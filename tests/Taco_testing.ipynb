```python
import sys
sys.path.append('../')
from tacobox import Taco
import cv2
```

## Taco Object

```python
mytaco = Taco()
```

## Input Image

```python
img = cv2.imread("../images/taco.png", cv2.IMREAD_GRAYSCALE)
```

```python
mytaco.visualize(img)
```

# Vertical TACO Examples

```python
# Random pixels
augmented = mytaco.apply_vertical_taco(img)
mytaco.visualize(augmented, 'Random Vertical')
# White pixels
augmented = mytaco.apply_vertical_taco(img, 'white')
mytaco.visualize(augmented, 'White Vertical')
# Black pixels
augmented = mytaco.apply_vertical_taco(img, 'black')
mytaco.visualize(augmented, 'Black Vertical')
# Mean Pixels
augmented = mytaco.apply_vertical_taco(img, 'mean')
mytaco.visualize(augmented, 'Mean Vertical')
```

# Horizontal TACO Examples

```python
# Random pixels
augmented = mytaco.apply_horizontal_taco(img)
mytaco.visualize(augmented, 'Random Horizontal')
# White pixels
augmented = mytaco.apply_horizontal_taco(img, 'white')
mytaco.visualize(augmented, 'White Horizontal')
# Black pixels
augmented = mytaco.apply_horizontal_taco(img, 'black')
mytaco.visualize(augmented, 'Black Horizontal')
# Mean Pixels
augmented = mytaco.apply_horizontal_taco(img, 'mean')
mytaco.visualize(augmented, 'Mean Horizontal')
```

# Both(Vertical + Horizontal) TACO 

```python
# Random pixels
augmented = mytaco.apply_taco(img)
mytaco.visualize(augmented, 'Random (V+H)')
# White pixels
augmented = mytaco.apply_taco(img, 'white')
mytaco.visualize(augmented, 'White (V+H)')
# Black pixels
augmented = mytaco.apply_taco(img, 'black')
mytaco.visualize(augmented, 'Black (V+H)')
# Mean Pixels
augmented = mytaco.apply_taco(img, 'mean')
mytaco.visualize(augmented, 'Mean (V+H)')
```
